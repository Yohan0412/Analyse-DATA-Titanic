```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

```

```python
df = pd.read_csv("titanic.csv")
df.head()
```

```python
df.describe()



```

```python
df = df.dropna(subset=["age"])
```

```python
df['survived'].value_counts().plot(kind="bar")
plt.title("Répartition des survivants")
plt.xticks([0, 1], ["Non", "Oui"])
plt.ylabel("Nombre de passagers")
plt.show()

pourcentage_survie = df['survived'].value_counts(normalize=True) * 100
print(pourcentage_survie)
```

Taux de survie du Titanic : 

Il y a 38% de taux de survivabilité

```python
sns.countplot(x='sex', hue='survived', data=df)
plt.title("Survie selon le sexe")
plt.xlabel("Sexe")
plt.ylabel("Nombre de passagers")
plt.legend(["Non survivants", "Survivants"])
plt.show()

pourcentage_survie_par_sexe = df.groupby('sex')['survived'].mean() * 100
print(pourcentage_survie_par_sexe)
```

# Taux de survie entre les hommes et les femmes :
On voit très bien ici que les femmes ont un taux de survie bien supérieur à celui des hommes. Cela peut s'expliquer par le fait que, pendant le sauvetage, les hommes se sont sacrifiés et ont laissé les femmes et les enfants monter dans les canots de sauvetage.

```python
df['age'] = pd.to_numeric(df['age'], errors='coerce')
df['age'] = df['age'].fillna(df['age'].mean())
df['age'] = df['age'].round().astype(int)
df = df.dropna(subset=['age'])
plt.figure(figsize=(15, 6)) 
sns.histplot(data=df, x="age", hue="survived", bins=20, kde=False, multiple="dodge")
plt.title("Répartition des âges selon la survie")
plt.xlabel("Âge")
plt.ylabel("Nombre")
plt.show()

```

# Taux de survivabilité du Titanic selon l'âge :
On voit que les enfants de moins de 10 ans ont un bon taux de survie par rapport aux autres tranches d’âge. Cela s'explique également par le fait que les hommes ont cédé leur place afin de sauver les femmes et les enfants

```python
sns.countplot(x='pclass', hue='survived', data=df)
plt.title("Survie selon la classe")
plt.xlabel("Classe")
plt.ylabel("Nombre de passagers")
plt.show()
```

# Survivabilité selon les Classe :
Ici, on voit que la première classe a un taux de survie plus élevé que les autres classes. Cela peut s'expliquer par le fait que la 1ʳᵉ classe était plus proche des canots de sauvetage, ce qui les rendait plus facilement accessibles au moment où ils étaient nécessaires.

```python

sns.boxplot(x="survived", y="fare", data=df)
plt.title("Prix du billet selon la survie")
plt.xticks([0, 1], ["Non survivants", "Survivants"])
plt.ylabel("Prix du billet (Fare)")
plt.show()
```
